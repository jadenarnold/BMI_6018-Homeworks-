```python
# Import Data Frame from 
#https://archive.ics.uci.edu/dataset/14/breast+cancer

from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
breast_cancer = fetch_ucirepo(id=14) 
  
# data (as pandas dataframes) 
X = breast_cancer.data.features 
y = breast_cancer.data.targets 
  
# metadata 
print(breast_cancer.metadata) 
  
# variable information 
print(breast_cancer.variables) 

print(breast_cancer)
#convert to data frame 
df= pd.DataFrame(data=X, columns=breast_cancer.variables['name']) 
df['Class']= y
print(df)
```

```python
#Part 3 
"""
Code to demonstrate call the library 
"""

#find all left 

# Filter rows where the 'breast' column contains the word' left' 
# so that we capture all the instances
df

# Regular expression pattern to find 'left' in 'breast' column
#case insensitive separates outs to more categories 
pattern = r'\bleft\b'
left_instances = df[df['breast'].str.contains(pattern, flags=re.IGNORECASE, na=False)]
print(left_instances[['breast']])

"""
This is my thoughts about this- I had to specify multiple ways this would be input,
if this was a more complex category this wouldn't have worked in the same way 
this library has particular uses, but definitely could be useful 
"""
```

```python
#spilt premeno & after post 
# This is another example- where very are going to split the data frame by premenopause and post-menopause 

df

# Applying regular expressions to split 'menopause' column
df[['premeno', 'ge40']] = df['menopause'].str.extract(r'(\bpremeno\b)|(\bge40\b)', flags=re.IGNORECASE)

# Drop the original 'menopause column
df.drop(columns=['menopause'], inplace=True)
print(df)
```
